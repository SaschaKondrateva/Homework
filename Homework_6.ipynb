```python
import string
read_text = read_text.translate(str.maketrans('','', string.punctuation))   
read_text = read_text.lower() 
read_text = read_text.split() 
print(read_text)
```

```python
result1 = sum([len(word) for word in read_text]) / len(read_text)
print(result1)
```
